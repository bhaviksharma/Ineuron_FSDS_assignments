# Q5
5. What method would you call to get Image object for a 100×100 image, excluding the lower-left quarter of it?

Ans.



```python
img1 = img.resize((100,100))
width, height = img1.size

left, top, right, bottom = (0, 0, width/2, height/2) #coordinates calculated from top left corner
img_crop1 = img1.crop((left, top, right, bottom)) # 1st quadrant

left, top, right, bottom = (width/2, 0, width, height/2) #coordinates calculated from top left corner
img_crop2 = img1.crop((left, top, right, bottom)) # 2nd quadrant

left, top, right, bottom = (width/2, height/2, width, height) #coordinates calculated from top left corner
img_crop4 = img1.crop((left, top, right, bottom)) # 4th quadrant
```

```python
img1
```

```python
img_crop1
```

```python
img_crop2
```

```python
img_crop4
```

```python
new_img = Image.new('RGB', img1.size, (250,250,250))
width, height = new_img.size
new_img
```

```python
new_img.paste(img_crop1, (0,0))
new_img.paste(img_crop2, (int(width/2), 0))
new_img.paste(img_crop4, (int(width/2), int(height/2)))
print("lower left quarter is excluded")
new_img

```
